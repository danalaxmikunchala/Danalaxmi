```python
import pandas as pd

df = pd.read_csv("orders.csv")

gold_df = df[df['membership_type'] == 'Gold']
city_revenue = gold_df.groupby('city')['total_amount'].sum().sort_values(ascending=False)
print(city_revenue)

cuisine_avg = df.groupby('cuisine')['total_amount'].mean().sort_values(ascending=False)
print(cuisine_avg)

user_total = df.groupby('user_id')['total_amount'].sum()
distinct_users = user_total[user_total > 1000].count()
print(distinct_users)

bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0-3.5', '3.6-4.0', '4.1-4.5', '4.6-5.0']
df['rating_range'] = pd.cut(df['rating'], bins=bins, labels=labels, include_lowest=True)
rating_revenue = df.groupby('rating_range')['total_amount'].sum().sort_values(ascending=False)
print(rating_revenue)

gold_city_avg = gold_df.groupby('city')['total_amount'].mean().sort_values(ascending=False)
print(gold_city_avg)

cuisine_restaurants = df.groupby('cuisine')['restaurant_name'].nunique()
cuisine_revenue = df.groupby('cuisine')['total_amount'].sum()
result = pd.DataFrame({'num_restaurants': cuisine_restaurants, 'total_revenue': cuisine_revenue})
print(result.sort_values(by=['num_restaurants','total_revenue'], ascending=[True, False]))

gold_orders = df[df['membership_type'] == 'Gold'].shape[0]
total_orders = df.shape[0]
percentage = round((gold_orders / total_orders) * 100)
print(f"{percentage}%")

restaurant_stats = df.groupby('restaurant_name').agg({'total_amount':['mean','count']})
restaurant_stats.columns = ['avg_order_value','total_orders']
result = restaurant_stats[restaurant_stats['total_orders'] < 20].sort_values(by='avg_order_value', ascending=False)
print(result)

combo_revenue = df.groupby(['membership_type','cuisine'])['total_amount'].sum().sort_values(ascending=False)
print(combo_revenue)

df['order_date'] = pd.to_datetime(df['order_date'])
df['quarter'] = df['order_date'].dt.quarter
quarter_revenue = df.groupby('quarter')['total_amount'].sum().sort_values(ascending=False)
print(quarter_revenue)
```
